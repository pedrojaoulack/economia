# Vamos ver como extrair dados financeiros de uma planilha e como visualizá-los por setor da economia:

```python
#Bibliotecas Necessárias:
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import yfinance as yf
import time
import random
```

```python
dados = pd.read_excel('dados_B3.xlsx', sheet_name=0)
dados_Ibov_amplo = pd.read_excel('indice_Amplo_Ibov.xlsx', sheet_name=0)
```

```python
print(dados)
print(dados_Ibov_amplo)
```

### Plotando o gráfico nestedpie circular:

```python
fig = px.sunburst(dados, path=['Setor', 'Sub-Setor', 'Segmento', 'Nome', 'Código'])
fig.update_layout(
    autosize=False,
    width=800,
    height=800,
    )
fig.show()

fig = px.sunburst(dados_Ibov_amplo, path=['Setor', 'Sub_Setor', 'Ação', 'Ticker'], values='Part. (%)')
fig.update_layout(
    autosize=False,
    width=800,
    height=800,
    )
fig.show()

```

```python
fig.write_html('gráfico_circular.html')
```

### plotando o gráfico nestedpie retangular:

```python
fig = px.treemap(dados, path=['Setor', 'Sub-Setor', 'Segmento', 'Nome'])
fig.update_layout(
    autosize=False,
    width=700,
    height=700,
    )
fig.show()

fig = px.treemap(dados_Ibov_amplo, path=['Setor', 'Sub_Setor', 'Ação', 'Ticker'], values='Part. (%)')
fig.update_layout(
    autosize=False,
    width=700,
    height=700,
    )
fig.show()
```

```python
fig.write_html('gráfico_retangular.html')
```
