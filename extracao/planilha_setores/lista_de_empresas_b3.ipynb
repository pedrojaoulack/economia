# Vamos ver como extrair dados financeiros de uma planilha e como visualizá-los:

```python
#Bibliotecas Necessárias:
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dados = pd.read_excel('dados_B3.xlsx', sheet_name=1)
```

```python
dados
```

### Plotando o gráfico:

```python
fig = px.sunburst(dados, path=['Setor', 'Sub-Setor', 'Segmento', 'Nome'])
fig.update_layout(
    autosize=False,
    width=900,
    height=900,
    )
fig.show()
```
