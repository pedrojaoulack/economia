# Como obter o gráfico do preço de fechamento ajustado (diário) de uma ação:

```
#Bibliotecas Necessárias:
import matplotlib.pyplot as plt
import pandas_datareader as web
import pandas as pd
```

## Como adquirir o ticker da empresa:
### vá no yahoo finance e veja o código da ação pesquisada na URL do site: https://finance.yahoo.com/quote/PETR3.SA?p=PETR3.SA&.tsrc=fin-srch
### no caso o ticker da ação vem depois do 'p=' , logo o ticker da petrobras será petr3.sa

```
#Variáveis base para obter os dados da ação:

#ticker da ação ordinária da petrobras:
ticker = "petr3.sa"
#se deixar a data inicial em branco ele irá buscar a data mais antiga disponível
data_inicial = "20/01/2015"
#se deixar a data final em branco ele irá buscar a data mais recente disponível
data_final = "20/01/2020"
```

```
#Obtendo os dados da ação:
acao = web.DataReader(ticker, 'yahoo', start=data_inicial, end=data_final)
```

```
plt.figure(figsize=(8,4))
plt.title(ticker)
plt.xlabel("de " + data_inicial + " até " + data_final)
plt.ylabel("Preço de fechamento ajustado")
plt.legend()
plt.plot(acao['Adj Close'], color='purple')
```
