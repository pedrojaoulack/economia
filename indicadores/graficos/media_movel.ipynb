# Vamos ver como obter a média móvel de uma ação e como projetar a mesma em um gráfico:

```python
#Bibliotecas Necessárias:
import matplotlib.pyplot as plt
import pandas_datareader as web
import pandas as pd
```

```python
#Variáveis base para obter os dados da ação:

#ticker da empresa Saudi Aramco, a maior empresa de petróleo do mundo, retirado do yahoofinance:
ticker="2222.sr"
#se deixar a data inicial em branco ele irá buscar a data mais antiga disponível
data_inicial = "01/01/2020"
#se deixar a data final em branco ele irá buscar a data mais recente disponível
data_final = "01/01/2021"
```

```python
#Obtendo os dados da ação:
acao = web.DataReader(ticker, 'yahoo', start=data_inicial, end=data_final)
```

### Vamos criar uma função da média móvel:

```python
periodos=20
def Media_Movel(periodos):
    acao['Média Móvel de ' + str(periodos) + 'p'] = acao['Adj Close'].rolling(periodos).mean()
```

### Vamos chamar a função Media_Movel e criar um gráfico comparando a média móvel com o preço da ação:

```python
Media_Movel(periodos=periodos)
```

```python
#Criando e customizando o gráfico
plt.figure(figsize=(8,4))
plt.title('Saudi Aramco')
plt.xlabel("de " + data_inicial + " até " + data_final)
plt.ylabel('Média Móvel de ' + str(periodos) + 'p')
plt.plot(acao['Adj Close'], color='black', label='Adj Close')
plt.plot(acao['Média Móvel de ' + str(periodos) + 'p'], color='purple', label='Média Móvel de ' + str(periodos) + 'p')
plt.legend()
```
